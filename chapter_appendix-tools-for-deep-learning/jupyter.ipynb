```python
# loading libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# uploading the file
ipl = pd.read_csv("matches.csv")
```

```python
# first 5 rows
ipl.head()
```

```python
# no. of rows and cols
ipl.shape
```

```python
# getting the frequency of most man of the match award
ipl.player_of_match.value_counts()
```

```python
# getting the top 10 from the above data
ipl.player_of_match.value_counts()[0:10]
```

```python
list(ipl.player_of_match.value_counts()[0:7].keys())
```

```python
# bar-graph for number of times player become man of the match
plt.figure(figsize=(10,5))
plt.bar(list(ipl.player_of_match.value_counts()[0:5].keys()), list(ipl.player_of_match.value_counts()[0:5]))
plt.show()
```

```python
# getting the frequency of result column
ipl.result.value_counts()
```

```python
# finding out no.of toss wins w.r.t each team
ipl.toss_winner.value_counts()[0:5]
```

```python
# getting the record where a team won batting first
batting_first = ipl[(ipl.target_runs)!=0]
```

```python
# getting the first 5 record of the above
batting_first.head()
```

```python
# making a histogram
plt.figure(figsize=(7,5))
plt.hist(batting_first.target_runs, bins=[50, 100, 150, 200, 250, 300])
plt.grid(True)
plt.title("winner by run")
plt.xlabel("runs")
plt.show()
```

```python
# finding out no. of wins w.r.t each team after batting first
batting_first.winner.value_counts()
```

```python
# making a bar-plot for first 3 teams with most run after batting first
plt.figure(figsize=(6,5))
plt.bar(list(batting_first.winner.value_counts()[0:3].keys()), list(batting_first.winner.value_counts()[0:3]), color=["green", "orange", "blue"])
plt.title("winner team after batting first")
plt.xlabel("teams")
plt.ylabel("counts")
plt.show()
```

```python
# making pie chart of the above
# 1st--> numerical data, 2nd---> categorical data
plt.figure(figsize=(6,5))
plt.pie(list(batting_first.winner.value_counts()[0:10]), labels= list(batting_first.winner.value_counts()[0:10].keys()), autopct="%0.1f%%")
plt.title("winner team after batting first")
plt.show()
```

```python
# finding out no. of wins w.r.t each team after batting second
batting_second = ipl[ipl.target_overs!=0]
```

```python
# getting the 1st 5
batting_second.head()
```

```python
# making a histogram for frequency of win w.r.t no. of wickets
plt.figure(figsize=(6,5))
plt.hist(list(batting_second.target_overs), bins=30)
plt.show()
```

```python
# finding out no. of wins w.r.t each team after batting second
batting_second.winner.value_counts()
```
